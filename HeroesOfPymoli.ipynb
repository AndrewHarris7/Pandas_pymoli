## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python


#Age Demographics
#Drop Duplicates
age_refined = purchase_data2.drop_duplicates("SN")

#The below each broken into bins of 4 years (i.e. <10, 10-14, 15-19, etc.)
bins = [0,9,14,19,24,29,34,39,100]
groups = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#Create a  new column for age groups and then groupby Age Groups
age_refined["Age Groups"] = pd.cut(age_refined["Age"], bins, labels=groups)
new_age = age_refined.groupby(["Age Groups"])


#Purchase Count
age_purchase = age_refined["Age Groups"].value_counts()

#Percentage of Users
age_percentage = age_purchase / player_count

age_demographics = pd.DataFrame({"Purchase Count": age_purchase,"Percentage of Players":age_percentage, })

age_demographics["Percentage of Players"] = age_demographics["Percentage of Players"].map("{:,.2%}".format) 

age_demographics = age_demographics.reindex(["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"])

age_demographics
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
grouped_id = purchase_data.set_index(["Item ID", "Item Name"])

grouped_id = grouped_id.groupby(level=["Item ID", "Item Name"])

#Find total spent per user
total_price_id = grouped_id.sum()["Price"]

#Find avg spent per user
avg_price_id = grouped_id.mean()["Price"]

#Find purchase count per user
count_id = grouped_id.count()["Price"]


#Create new dataframe
items_df = pd.DataFrame({"Count":count_id,"Average Purchase Price":avg_price_id,
                            "Total Purchase Price": total_price_id,})

#Sort by total purchase price
revised_items = items_df.sort_values("Count",ascending=False)

revised_items["Average Purchase Price"] = revised_items["Average Purchase Price"].map("${:,.2f}".format) 
revised_items["Total Purchase Price"] = revised_items["Total Purchase Price"].map("${:,.2f}".format) 

#Reorder Columns
revised_items = revised_items[["Count", "Average Purchase Price", "Total Purchase Price"]]


#Display top 5
revised_items.head(5)


```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
grouped_id = purchase_data.set_index(["Item ID", "Item Name"])

grouped_id = grouped_id.groupby(level=["Item ID", "Item Name"])

#Find total spent per user
total_price_id = grouped_id.sum()["Price"]

#Find avg spent per user
avg_price_id = grouped_id.mean()["Price"]

#Find purchase count per user
count_id = grouped_id.count()["Price"]

#Create new dataframe
items_df = pd.DataFrame({ "Count":count_id,"Average Purchase Price":avg_price_id,
                        "Total Purchase Price": total_price_id })


#Sort by total purchase price
profit_items = items_df.sort_values("Total Purchase Price",ascending=False)

profit_items["Average Purchase Price"] = profit_items["Average Purchase Price"].map("${:,.2f}".format) 
profit_items["Total Purchase Price"] = profit_items["Total Purchase Price"].map("${:,.2f}".format) 

#Reorder Columns
profit_items = profit_items[["Count", "Average Purchase Price", "Total Purchase Price"]]

#Display top 5
profit_items.head(5)
```
